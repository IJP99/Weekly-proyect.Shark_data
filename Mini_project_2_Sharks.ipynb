```python
# function which store every word with surf in surfing in order to minimize the outputs

def categorize_activity(Activity):
    Activity = str(Activity).lower()
    if "surf" in Activity or "boarding" in Activity:
        return "surfing"
    elif "swim" in Activity or "snorkeling" in Activity:
        return "swiming"
    elif "fish" in Activity:
        return "fishing"
    else:
        return "other activities"
        return Activity
```

```python
states['Activity'] = states['Activity'].apply(categorize_activity)

states["Activity"].unique()
```

```python
states["Sex"].unique()
```

```python
#dealing with the misspelled inputs inside of the Sex column

def clean_spaces_2(states, Sex):
    states["Sex"] = states["Sex"].str.strip()
    return states
```

```python
states = clean_spaces_2(states, "Sex")
states["Sex"].unique()
```

```python
states["Age"].unique()
```

```python
#dealing with inputs that have special characters

import re
pattern = r"[!\¡]"

states['Age'] = states['Age'].replace(pattern, '', regex=True)
```

```python
states["Age"].unique()
```

```python
#non valid input

states = states.drop(states[states['Age'] == '22, 57, 31'].index)
```

```python
states["Age"].unique()
```

```python
#Function that turns the value "teen" into a int and the convert the str values to float

def clean_and_convert_to_float(value):
    if isinstance(value, str) and value.lower() == "teen":
        return 10
    if isinstance(value, str) and value[-1].isdigit():
        return float(value)
    else:
        return float(value[:-1]) if isinstance(value, str) else value

```

```python
states["Age"] = states["Age"].apply(clean_and_convert_to_float)
```

```python
states["Age"].unique()
```

```python
#filling the Nan values of the column Age with the mean just to not affect the total variables of the column

states["Age"] = states["Age"].fillna(states["Age"].mean())
```

```python
states["Type"].value_counts()
```

```python
states3=states.copy()
```

```python
#categorize type in 3 types

def categorize_type(Type):
    if isinstance(Type, str):
        if Type.lower().strip() in ["unprovoked", "provoked"]:
            return Type
        elif "invalid" in Type.lower():
            return "others reasons"
        elif "watercraft" in Type.lower():
            return "others reasons"
        elif "questionable" in Type.lower():
            return "others reasons"
        elif "under investigation" in Type.lower():
            return "others reasons"
        else:
            return Type
```

```python
states3["Type"]=states3["Type"].apply(categorize_type)
```
